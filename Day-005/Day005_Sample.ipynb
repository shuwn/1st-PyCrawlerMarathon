## 下載檔案

```python
import requests
# 得到一個 Response 物件叫做r，可以得到這個物件裡面的資訊像是status_code, url, text(網頁內容)等。
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')

# http 請求的返回狀態，2XX 表示連線成功，3XX 表示跳轉 ,4XX 客戶端錯誤 , 500 伺服器錯誤
print(r.status_code)

# 請求返回的網址
print(r.url)

# 請求返回的編碼
print(r.encoding)

# 將擷取的資料轉換為 text 文件
# r.text：這個是 requests 自動識別獲取的內容進行對應的解碼，有可能得到亂碼
# 可能需要解碼（比如decode utf-8）才能看到在瀏覽器上面顯示的內容。
response = r.text

response
```

```python
import json
data = json.loads(response)

for d in data:
    print(d['title'])
```
