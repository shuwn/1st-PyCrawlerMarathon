## 作業目標

根據範例 ，完成以下問題：

* ① 印出最新文章的「作者」「標題」「時間」
* ② 印出第一頁所有文章的「作者」「標題」「時間」


```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime

url = "https://www.ptt.cc/bbs/Gossiping/index.html"

r = requests.get(url, cookies={'over18': '1'})
soup = BeautifulSoup(r.text, 'xml')
# print(soup)
```

### ① 印出最新文章的「作者」「標題」「時間」

```python
# 原始文章排序
newPapers = []
for d in soup.find_all(class_='title'):
    try:
        newPaper = {}
        newPaper['標題'] = d.text.replace('\t', ' ').replace('\n', ' ')
        r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href'], cookies={'over18': '1'}).text, 'xml')
        newPaper['作者'] = r.find(class_='article-meta-value').text
        newPaper['時間'] = r.find_all(class_='article-meta-value')[-1].text
        newPapers.append(newPaper)
    except:
        continue
        
newPapers
```

```python
# 依時間排列
newPapers = []
for d in soup.find_all(class_='title'):
    try:
        newPaper = {}
        newPaper['標題'] = d.text.replace('\t', ' ').replace('\n', ' ')
        r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href'], cookies={'over18': '1'}).text, 'xml')
        newPaper['作者'] = r.find(class_='article-meta-value').text
        newPaper['時間'] = r.find_all(class_='article-meta-value')[-1].text
        newPapers.append(newPaper)
    except:
        continue

# sort 與 sorted 區別：
# sort 是應用在 list 上的方法，sorted 可以對所有可迭代的對象進行排序操作。
# list 的 sort 方法返回的是對已經存在的列表進行操作，無返回值，而內建函數 sorted 方法返回的是一個新的 list，而不是在原來的基礎上進行的操作。
# sort 僅針對 list，不支援字典
# lambda 實際上就是匿名函數，以參數 key 傳入 lambda，指定要根據哪個元素進行排序
# key=lambda 元素：元素 [欄位索引]
# 例如：想對元素第二個欄位排序，則
# key=lambda y: y [1] 備註：這裡 y 可以是任意字母，等同 key=lambda x: x [1]
# 根據 age 年齡排序：
# print(sorted(data, key=lambda x: x['age']))
# 先排 age 年齡、再排 name 名字：
# print(sorted(data, key=lambda x: (x['age'], x['name'])))
newPapers = sorted(newPapers, key=lambda x: x['時間']) 
newPapers
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
# 方法一
for d in soup.find_all(class_='title'):
    print(d.text.replace('\t', ' ').replace('\n', ' '))
    try:
        r = BeautifulSoup(requests.get('https://www.ptt.cc/'+d.find('a')['href'], cookies={'over18': '1'}).text, 'xml')
        # print(r)
        print('作者: '+r.find(class_='article-meta-value').text)
        print('時間: '+r.find_all(class_='article-meta-value')[3].text)
        # print('時間: '+r.find_all(class_='article-meta-value')[-1].text)
    except:
        continue
```

```python
# 方法二
for d in soup.find_all(class_='r-ent'):
    print('標題： '+d.find(class_='title').text.replace('\t', ' ').replace('\n', ' '))
    print('作者： '+d.find(class_='author').text.replace('\t', ' ').replace('\n', ' '))
    print('時間： '+d.find(class_='date').text.replace('\t', ' ').replace('\n', ' '))
```
