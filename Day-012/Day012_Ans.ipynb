### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

res = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text
    title = d.find_all('a')[-1].text
    # tag = d.find(class_='tag').text
    tag = d.select('.tag')[0].text
    # dict.setdefault(key, default=None)，如果鍵不存在於字典中，將會添加鍵並將值設為默認值。
    res.setdefault(tag, [])
    res[tag].append({
        'Title': title, 'Date': date
    })
    
print(res)
```

### ③ 哪一個類別的文章最多

```python
# ③ 哪一個類別的文章最多
data = []
for i in res:
    d = {}
    d['tag'] = i
    d['count'] = len(res[i])
    data.append(d)
    print(i, len(res[i]))

for e in sorted(data, key=lambda d: d['count'])[::-1]:
    print(e)
    # sorted 可以對所有可迭代的物件進行排序操作。
    # sorted 方法返回的是一個新的 list，而不是在原來的基礎上進行的操作。
```
