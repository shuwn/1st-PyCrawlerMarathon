```python
import requests
import time
from bs4 import BeautifulSoup
```

### ① 取出今天所有的發文

```python
url = 'https://www.ettoday.net/news/news-list.htm'

r = requests.get(url)

soup = BeautifulSoup(r.text, "lxml")

# print(soup)
for d in soup.find(class_="part_list_2").find_all("h3"):
    print(d.find(class_="date").text, d.find(class_="tag").text, d.find_all("a")[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
res = {}
for d in soup.find(class_="part_list_2").find_all("h3"):
    date = d.find(class_="date").text
    tag = d.find(class_='tag').text
    title = d.find_all("a")[-1].text
    # tag = d.select('.tag')[0].text
    # soup.select() 是利用 CSS selector 的方式做篩選
    # 使用 select () 函數找出所有 tag 元素，select()是使用 CSS 選擇器的語法找到 tag，故需要添加 .（依標籤類別）
    # dict.setdefault(key, default=None)，如果 Key 不存在於字典中，將會添加該 Key 並將 default 的值設為該 Key 的默認值
    # 如果 Key 存在於字典中，將讀出該 Key 原來對應的值，default 的值不會覆蓋原來已經存在的 Key 的值
    # key----要查找的鍵
    # default-----查找的鍵不存在時用於設置的默認值
    # https://blog.csdn.net/q389797999/article/details/81239499
    res.setdefault(tag, [])
    # 【】 為列表
    # 將 tag 對應一個空列表
    # 依照迴圈，將對應的 Tag、Title、Date 添加到空列表中
    # 如果列表中已有對應的 Key，會將 Title、Date 添加到 Key 原來對應的列表之後 
    res[tag].append({
     "Title" : title, "Date" : date
    })
print(res)
```
