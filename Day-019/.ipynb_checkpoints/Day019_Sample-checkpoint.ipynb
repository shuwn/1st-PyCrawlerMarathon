## 動態模擬瀏覽器行為

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
```

```python
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://airtw.epa.gov.tw/")
```

![](https://imgur.com/LwGGT8c.jpg)

```python
selectSite = Select(browser.find_element_by_id("ddl_county"))
selectSite.select_by_value('Taipei')
selectYear = Select(browser.find_element_by_id("ddl_site"))
selectYear.select_by_value('5')
```

```python
# 新版不需要點選查詢
# browser.find_element_by_id('ctl05_btnQuery').click()
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
# html_source
```

```python
soup = BeautifulSoup(html_source, 'html.parser')
print(soup)
# 新版內容錯誤
# table = soup.find('AQI_title', class_='TABLE_G')
# print(table)
```
